# Mean & Std.dev.

```python
print("Mean of survival")
print(np.mean(haber_survival["nodes"]))
print("---------------------------------------")
print("Mean with outlier of survival")
print(np.mean(np.append(haber_survival["nodes"],50)))
print("---------------------------------------")
print("mean of not survival")
print(np.mean(haber_not_survival["nodes"]))
print("***************************************")
print("std of survival")
print(np.std(haber_survival["nodes"]))
print("***************************************")
print("std of not survival")
print(np.std(haber_not_survival["nodes"]))
```

```python
print("Median of survival")
print(np.median(haber_survival["nodes"]))
print("***************************************")
print("Median with outlier survival")
print(np.median(np.append(haber_survival["nodes"],50)))
print("***************************************")
print("Median of not survival")
print(np.median(haber_not_survival["nodes"]))
```

# Percentile, Quantiles,IQR,MAD

```python
print("Quantiles")
print(np.percentile(haber_survival["nodes"],np.arange(0,100,25)))
print(np.percentile(haber_not_survival["nodes"],np.arange(0,100,25)))
print("***************************************")
print("90th percentile")
print(np.percentile(haber_survival["nodes"],90))
print(np.percentile(haber_not_survival["nodes"],90))
```

# MAD

```python
from statsmodels import robust
print(robust.mad(haber_survival["nodes"]))
print(robust.mad(haber_not_survival["nodes"]))
```
